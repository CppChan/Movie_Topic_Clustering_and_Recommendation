## Tune param(num of hidden layers)

```python
import tensorflow as tf
from train_ import train
```

```python
test_layer={}
for layer in range(3,22,3):
    tr = train(num_layer=layer, reg = 'L2')
    test_layer[layer]=tr.train(None)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

X = [3,6,9,12,15,18,21]
Y = [[0]*7 for i in range(7)]

for j in range(7):
    for i in range(len(X)):
        Y[j][i] = test_layer[X[i]][j][2]
    
plt.figure(figsize=(9, 6), dpi=80)
ax = plt.subplot(111)
for j in range(len(Y)-1):
    ax.plot(X,Y[j],label='num of epoch='+str(j*5),linewidth=3)

ax.legend(numpoints=1, loc = 1)
plt.xlabel('Num of Hidden Layers')
plt.ylabel('RMSE')
plt.title('When num of epoch is 20, relationship between num of layers and RMSE')
plt.show()
```

#### We can conclude that more hidden layers is good to our model performance, which means RMSE will decrease. However, we can also see that high num of epoch(>15) and high num of hidden layers(>18) is easy to lead to overfitting. Therefore, we can conclude that the best parameter of num_layers and num_epoch is around (12~16) and (5~10)

## Tune param(activation func)

```python
import tensorflow as tf
from train_ import train
```

```python
tr = train(num_layer=6, reg = 'L2', act_func='Relu')
X = [0,5,10,15,20,25,30]
Y_sigmoid = tr.train(None)
```

```python
Y_relu = test_layer[6]
```

```python
temp1,temp2 = [],[]
for i in range(len(Y_relu)):
    temp1.append(Y_relu[i][2])
    temp2.append(Y_sigmoid[i][2])
Y_relu,Y_sigmoid = temp1,temp2
```

```python
plt.figure(figsize=(9, 6), dpi=80)
ax = plt.subplot(111)
ax.plot(X,Y_sigmoid,label='Activation_func=sigmoid',linewidth=5)
ax.plot(X,Y_relu,label='Activation_func=relu',linewidth=5)
ax.legend(numpoints=1, loc = 1)
plt.xlabel('Epoch')
plt.ylabel('RMSE')
plt.title('Different performance of activation function when num of hidden layer is 6')
plt.show()
```

#### We can see that different activation function is not the main factor to affect model performance

## Tune param(regularization term)

```python
import tensorflow as tf
from train_ import train
```

```python
tr = train(num_layer=6, reg = 'L1')
X = [0,5,10,15,20,25,30]
Y_L1 = tr.train(None)
```

```python
Y_L2 = test_layer[6]
```

```python
temp1,temp2 = [],[]
for i in range(len(Y_L1)):
    temp1.append(Y_L1[i][2])
    temp2.append(Y_L2[i][2])
Y_L1,Y_L2= temp1,temp2
```

```python
plt.figure(figsize=(9, 6), dpi=80)
ax = plt.subplot(111)
ax.plot(X,Y_L1,label='regularization=L1',linewidth=5)
ax.plot(X,Y_L2,label='regularization=L2',linewidth=5)
ax.legend(numpoints=1, loc = 1)
plt.xlabel('Epoch')
plt.ylabel('RMSE')
plt.title('Different performance of regularization term')
plt.show()
```

#### We can see that L2 regularization slightly performs better than L1 regularization
